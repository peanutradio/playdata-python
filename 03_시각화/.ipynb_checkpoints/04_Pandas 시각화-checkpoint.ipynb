## plot() 
- kind 매개변수에 지정한 값에 따라 다양한 그래프를 그릴 수 있다.
- kind : 그래프 종류 지정
    - 'line' : line plot (default)
    - 'bar' : vertical bar plot (수직)
    - 'barh' : horizontal bar plot (수평)
   
    - 'hist' : histogram
    - 'box' : boxplot
    - 'kde' : Kernel Density Estimation plot (부드러운 곡선)
    - 'pie' : pie plot
    - 'scatter' : scatter plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
s = pd.Series([100,70,90,130,50], index=['사과', '배', '귤','수박', '복숭아'] )   
s
```

```python
plt.figure(figsize=(4,3))
s.plot(kind='bar', title='과일수량', xlabel='과일종료', ylabel='개수')  #판다스
# plt.title('과일수량')   #matplotlib 으로 설정
# plt.xlabel('과일')
# plt.ylabel('수량')
plt.show()
```

matplotlib를 이용해 여러 설정들 가능

```python
s.plot.bar(title='과일수량', xlabel='과일종료', ylabel='개수', figsize=(6,5))

plt.show()
```

```python
s.plot.bar(title='과일수량', xlabel='과일종료', ylabel='개수', figsize=(4,3), rot=45)

plt.show()
```

## 막대 그래프
- index가 무슨 값인지를 가리키는 축으로 사용된다.

```python
tips = pd.read_csv('data/tips.csv')
tips.shape
```

```python
tips.head(3)
```

```python
cnt_result = tips['day'].value_counts()
cnt_result
```

```python
cnt_result.plot(kind='bar', title='요일별 테이블수', xlabel='요일', ylabel='테이블수', grid=True, rot=45)
plt.show()
```

```python
tips['smoker'].value_counts().plot(kind='barh', figsize=(5,3), title='흡연여부에 따른 테이블 수', 
                                  color=['g', 'r'])
plt.show()
```

- 두개의 분류 별로 그리기
    - 여러개의 컬럼일 경우 수평 누적 막대그래프를 그린다.


```python
agg_df = tips.pivot_table(index='smoker', columns='sex', values='tip', aggfunc='count')
agg_df                                                         #aggfunc='count' => 빈도수
```

```python
agg_df.plot(kind='bar', title='흡연여부/ 성별 테이블 수',xlabel='흡연여부', ylabel='테이블수')
plt.show()
```

```python
#요일/성별 손님의 총수(size) --> 중간 TODO -> 다시 해봐야!!
# result = tips.pivot_table(index='day', columns='sex', values='size', aggfunc='sum')
result = tips.pivot_table(index='sex', columns='day', values='size', aggfunc='sum')

result
```

```python
result.plot(kind='bar') # stacked=True)  #stacked - 그래프 탑 쌓기.
plt.legend(bbox_to_anchor=(1,1), loc='upper left', title='day')
plt.show()
```

```python
result.plot(kind='bar', subplots=True, figsize=(6,10))  #subplot으로 각각 그린다. 
plt.show()
```

## 파이차트

```python
# tips['day'].value_counts(normalize=True)
tips['day'].value_counts().plot(kind='pie', autopct='%.2f%%', explode=[0,0,0,0.2], shadow=True)
plt.show()
```

```python
tips.groupby('time')['tip'].mean().plot.pie(autopct='%.2f%%', shadow=True, explode=[0,0.2])
plt.show()
```

## 히스토그램, KDE(밀도그래프)

```python
# tips['tip'].plot(kind='hist', bins=20)   #bins = 쪼개는 것을 더 세분화.
tips['tip'].plot.hist(bins=20)    # 위와 동일한 코딩.
plt.show()
```

```python
tips['tip'].plot(kind='kde')   #kde = 곡선형태 그래프
plt.show()
```

## Boxplot (상자그래프)

```python
tips[['total_bill']].plot(kind='box', whis=3)
plt.show()
```

```python
tips[['tip', 'total_bill']].plot(kind='box',figsize=(5,4))
plt.show()
```

## scatter plot (산점도) 

```python
# scatter는 두가지의 값을 비교하는 표이다. 
# 그래서 columns가 2개 이상이어야 한다. 그래서 Series가 아니라 DataFrame을 이용해서 그리는 것이다.
#    (Series는 그릴수 없다.)
```

```python
tips.plot(kind='scatter', x='total_bill', y='tip')
plt.show()
```

```python
tips[['total_bill', 'tip']].corr()    # corr()==> 상관계수 확인
```
