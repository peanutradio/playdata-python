# 4. 파이차트 그리기
## 4.1 파이차트
- 각 범주(Category)가 데이터에서 차지하는 비율을 나타내는데 사용
- `pie(x, labels)` 이용
    - x: 값 (값들을 100을 기준으로 비율을 계산해 크기 설정)
    - labels : 값들의 label
    - autopct: 조각내에 표시될 비율의 문자열 형식. '%fmt문자' 
        - fmt문자: f(실수), d(정수), %% (%)
            

```python
import matplotlib.pyplot as plt  # matplotlib 설치시 혹은 터미널 시작시, 1번과 2번, 3번 명령어 입력으로 폰트깨짐 해결.(Mac 한정.)
from matplotlib import rc        # 1번
rc('font', family='AppleGothic') # 2번
plt.rcParams['axes.unicode_minus'] = False #3번
```

```python
import matplotlib.pyplot as plt
labels = ['사과', '배', '수박', '귤', '복숭아']
counts = [20,10,50,40,30]
plt.figure(figsize=(7,7))

#pie(수량, 라벨)
plt.pie(counts, labels=labels, 
       autopct='%.2f%%',   #소수점 이하 두자리수까지  / f=flout
       shadow=True,        #그림자
       explode=[0,0,0.2,0,0.2])  # 해당되는 값만 수치만큼 뛰워준다(강조하고 싶을 때 사용한다.)
plt.show()
```

```python
plt.figure(figsize=(6,6))

wedge, texts, autopcts = plt.pie(counts, 
                   #     labels=labels,       #pie chart 그릴 때, label을 빼고 legen() 설정
                         autopct='%.2f%%',   #소수점 이하 두자리수까지  / f=flout
                         shadow=True,        #그림자
                         explode=[0,0,0.2,0,0.2])  # 해당되는 값만 수치만큼 뛰워준다(강조하고 싶을 때 사용한다.)

# Label을 범례로 빼고자 할때.
plt.legend(wedge, labels,   #wedge > color
           bbox_to_anchor=(1,1), loc='upper left', 
           title='과일')
```

# 5 히스토그램(Histogram) 그리기
## 5.1 히스토그램

- 도수 분포표를 그래프로 나타낸 것.
    - 도수분포표: 연속형 자료를 특정 구간(bin)으로 나눠 그 빈도를 나타낸 표
        - 빈도 확인이나 **분포**를 볼때 사용한다.
    - X축: 계급(변수구간)의 대표값, Y축: 빈도수
- `hist(data [, bins=계급개수)` 메소드를 사용
    - data: 연속형 값들을 원소롷는 1차원 자료구조.

```python
data = np.random.normal(100,10,1000)   # 평균 : 100, 표준편차: 10 정규분포를 따르는 난수 1000개.
# data
v, cnt = np.unique(data, return_counts=True)   #고유값과 개수.
cnt.max()
```

```python
plt.hist(data,bins=30)
plt.show()
```

# 6. 상자그래프(Boxplot) 그리기
## 6.1 상자그래프란
- 연속형 값들의 사분위수, 극단값들을 이용해 값들의 분포를 시각화한다.
- boxplot(x, whis=1.5)
    - x: 데이터셋
    - whis: 극단치 계산시 사용할 값(기본 1.5)
        - IQR을 이용해 극단치 계산하기
            - IQR(Inter Quatile Range): $3분위수 - 1분위수$ 로 중간 50%값의 범위
            - 극단적으로 작은값: 1분위 - $IQR(3분위수-1분위수) * whis$   보다 작은 값들
            - 극단적으로 큰값: 3분위 + $IQR(3분위수-1분위수) * whis$  보다 큰 값들

```python
import numpy as np
data = np.random.randint(1,100,1000) # 1 ~ 99 사이의 난수 정수 1000개를 1차원 배열에 묶어서 반환.
data[:5] = [200,250,-50,-60,-80]
data.shape
```

```python
# plt.boxplot(data, whis=3)   #IQR x 1.5배  => IQR x 3배 로 늘림.
plt.boxplot(data, whis=0.1)   #IQR x 1.5배  => IQR x 0.1배 로 줄임 

plt.show()
```
