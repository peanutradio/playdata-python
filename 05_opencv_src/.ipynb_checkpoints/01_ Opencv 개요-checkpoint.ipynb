# Image란?(영상이란?)

## Image는 width X height X channel 의 pixels로 이루어진 matrix   </br>(실제 코딩 시 height X width X channel로 표기 한다.)

- width: 너비로 이미지의 가로 길이를 의미한다.
- height: 높이로 이미지의 세로 길이를 의미한다.

- pixel: image가 가지고 있는 값으로 width X height 개수만큼 존재한다. image를 표현하는 bit방식에 따라 값의 범위가 달라진다.

- 일반적인 Image는 unsigned integer 8bit로 표현
- 0 ~ 255 로 표현 / 0(검정) ~ 255(흰색)
- bit수가 높아질 수록 표현할 수 있는 색의 범위가 증가한다.

## 색공간 변환
- **`cv2.cvtColor(src, code)`**
    - image 의 color space를 변환한다. 
    - src: 변환시킬 이미지 (ndarray)
    - code
        - 변환시킬 색공간 타입 지정
        - cv2.COLOR_XXX**2**YYY 형태의 상수 지정(XXX를YYY로 변환)
            - cv2.COLOR_XXX2YYY 형태의 상수 지정(XXX를YYY로 변환)
            - cv2.COLOR_BGR2GRAY / cv2.COLOR_GRAY2BGR (BGR <-> GRAY)
            - cv2.COLOR_BGR2RGB  / cv2.COLOR_RGB2BGR  (BGR <-> RGB)
            - cv2.COLOR_BGR2HSV  / cv2.COLOR_HSV2BGR  (BGR <-> HSV)

> ### HSV
> - Hue: 색상, 색의 종류
> - Saturation: 채도. 색의 선명도
> - Value : 명도. 밝기

# 동영상 처리

## 동영상 읽기
- VideoCapture 클래스사용
    - 객체 생성
        - VideoCapture('동영상파일 경로'): 동영상파일
        - VideoCapture(웹캠 ID): 웹캠
- VideoCapture의 주요 메소드
    - isOpened(): bool
        - 입력 대상과 연결되었는지 여부반환
    - read(): (bool, img)
        - Frame 이미지로 읽기
        - 반환값
            - bool: 읽었는지 여부
            - img: 읽은 이미지(ndarray)

#### 웹캠

```python
import cv2

cap = cv2.VideoCapture(0)  # 정수: web cam 연결  => 0부터 1씩 증가하는 값을 사용.웹캠이 여러개인 경우 지정.
print(cap.isOpened())
while True:
    # 한 개 Frame을 읽기
    sucess, image = cap.read()
    if not sucess:
        print('프레임을 읽어오지 못함')  # 만약, sucess를 읽지 못한다면, 빠져나와서 종료
        break
    
    # 읽은 Frame을 출력
    image = cv2.flip(image, 1)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame', image)    # 양수: 수평반전 / 0: 수직반전, / 음수: 수평, 수직 반전
    if cv2.waitKey(1)  == 27:  # 27: esc key 누리면 종료.
        break
    
cap.release()  # 연결 닫기(해제)
cv2.destroyAllWindows()
cv2.waitKey(1)   
```

```python
# 웹캠으로 사진찍기 (video_cam_take_pic.py)

import cv2

cap = cv2.VideoCapture(0)                       # 0번 카메라 연결
if cap.isOpened() :
    while True:
        ret, frame = cap.read()                 # 카메라 프레임 읽기
        if ret:
            cv2.imshow('camera',frame)          # 프레임 화면에 표시
            if cv2.waitKey(1) != -1:            # 아무 키나 누르면
                cv2.imwrite('photo.jpg', frame) # 프레임을 'photo.jpg'에 저장
                break
        else:
            print('no frame!')
            break
else:
    print('no camera!')
cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)   
```

#### 동영상파일

- FPS (Frame Per Second) -  1초에 몇 Frame을 보여주는지 여부. 동영상 파일에 설정되어 있다. 
- 동영상 파일의 FPS에 맞춰서 (속도를 맞춰서) 플레이 하려면 한 Frame을 보여주고 FPS 맞는 시간간격만큼 기다려야 한다.
  - 기다리는 시간:  초단위/FPS

```python
# FPS 설정
cap_tmp = cv2.VideoCapture('images/wave.mp4')  #cv2.VideoCapture(0)
FPS = cap_tmp.get(cv2.CAP_PROP_FPS)
DELAY1 = 1/FPS # 1 Frame을 보여줄 시간이 어떻게 되는지
DELAY2 = 1000/FPS
print(FPS, f"{DELAY1}초, {DELAY2}밀리초")
cap_tmp.release()
```

```python
import cv2
#VideoCapture 객체 생성 -> 동영상 파일 경로를 문자열로 전달.
cap = cv2.VideoCapture("images/wave.mp4")  # 정수: web cam 연결 / 문자열: 동영상 파일 연결
print("연결성공여부:", cap.isOpened())

FPS = cap.get(cv2.CAP_PROP_FPS)
DELAY = int(1000/FPS)  #waitKey() 실수가 들어갈수 없어서 int값을 대기시간으로 받는다(정수)

while cap.isOpened():
    success, image = cap.read()   # 한 Frame 일기
    if not success:
        print('읽기 실패')
        break # 종료
        
    cv2.imshow('frame', image) # 색상으로 GRAY로도 바꿀수 있다. -> cv2.COLOR_BGR2GRAY)
    if cv2.waitKey(1)  == 27:  # esc로 빠져나오기.
        break

cap.release()  #연결 닫기(끊기)
cv2.destroyAllWindows()
cv2.waitKey(1)  
```

### 동영상 저장
- capture(read)한 이미지 프레임을 연속적으로 저장하면 동영상 저장이 된다.
- VideoWriter 객체를 이용해 저장
    - <span style="color:orange">**VideoWriter(filename, codec, fps, size)**</span>
        - filename: 저장경로
        - codec
            - VideoWriter_fourcc 이용
            
        - fps
            - FPS (Frame Per Second) - 초당 몇 프레임인지 지정
        - size
            - 저장할 frame 크기로 원본 동영상이나 웹캠의 width, height 순서로 넣는다.
    - `VideoWriter().write(img)`
        - Frame 저장

```python
def func(*args, **kwargs):   # *args: 가변인자   / **kwargs:dictionary로 준다.
    print(args)
    print(kwargs
         )
# func(1,2,3,4,5,'a','b','c','d')
a = [1,2,3,4,5]
b = dict(a=1, b=2, c=3)
func(a)
func(*a)
func('가나다라마바사')
func(*'가나다라마바사')
func('DIVXS')
func(*'DIVXS')  #=>영상 코덱 지정해줄때는 가변인자로 줘야 한다.  

func(*a, **b)
```

```python
# 웹캠에서 읽은 동영상을 파일로 저장
import cv2

cap = cv2.VideoCapture(0)

# 입력 영상에 size 지정
_, img = cap.read()
print(img.shape)
size = (img.shape[1], img.shape[0])  # width, height
print(size)

# Codec 지정
codec = cv2.VideoWriter_fourcc(*"MJPG")  # 코덱 지정.
# FPS 지정
FPS = cap.get(cv2.CAP_PROP_FPS)
#filepath 지정.
filepath = "webcam_output.avi"


#VideoWriter 객체 성성
writer = cv2.VideoWriter(filepath, codec, FPS, size)

while True:
    success, image = cap.read()
    if not success:
        print('캡쳐실패')
        break
        
    cv2.imshow('frame', image)
    writer.write(image)
    if cv2.waitKey(1) ==27:
        break
    
cap.release()
writer.release()
cv2.destroyAllWindows()
cv2.waitKey(1) 
```
